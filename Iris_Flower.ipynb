```python
import pandas as pd
import matplotlib.pyplot as plt

iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

```python
iris.head()
```

```python
iris.value_counts(iris.species)
```

三种花的平均花瓣长度各是多少？

```python
avg_iris_petal_length = iris.groupby("species")["petal_length"].mean().reset_index()
```

```python
iris.groupby("species")["petal_length"].mean().reset_index()
```

三种花的平均花萼宽度各是多少？

```python
iris.groupby("species")["sepal_width"].mean()
```

```python
setosa = iris[iris["species"] == "setosa"]
versicolor = iris[iris["species"] == "versicolor"]
virginica = iris[iris["species"] == "virginica"]
```

```python
plt.scatter(setosa["petal_length"],setosa["sepal_width"],label="Setosa",color="pink")
plt.scatter(versicolor["petal_length"],versicolor["sepal_width"],label="Versicolor",color="tan")
plt.scatter(virginica["petal_length"],virginica["sepal_width"],label="Virginica",color="green")
plt.title("Relationship between petal length and sepal width")
plt.xlabel("Petal Length")
plt.ylabel("Sepal Width")
plt.legend()
```

```python
plt.bar(avg_iris_petal_length["species"],avg_iris_petal_length["petal_length"])
plt.show()
```
